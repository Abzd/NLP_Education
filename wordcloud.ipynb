```python
###############################绘制词云######################################
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud,ImageColorGenerator
from scipy.misc import imread
```

```python
def wordcloud(data,back):
    data = ''.join([i for i in data if isinstance(i,str) if len(i)>4])
    
    wc = WordCloud(background_color="white",mask=back,max_words=100,
    font_path="C:/Windows/Fonts/STFANGSO.ttf").generate(data)

    plt.figure(figsize=(15,10))
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
```

```python
ted = pd.read_csv('D:/NLP_Education/data/domestic_tags.csv')
back = imread('D:/NLP_Education/timg.jpg')
wordcloud(ted['Content'],back)
```

```python
domestic = pd.read_csv('D:/NLP_Education/data/domestic_tags.csv')
```

```python
import re
def plot_user(data):
    df = pd.DataFrame()
    pattern = '网易(.*?)网友.*'
    df['User'] = data['User'].apply(lambda x:re.search(pattern,x).group(1))
    df['type'] = data['User'].apply(lambda x:x[-2:] if x[-2:]=='手机' else '电脑')
    way = df['type'].value_counts()
    
    x = [0,1]
    plt.bar(x,way)
    plt.xticks(x,['Mobile','PC'])
    
plot_user(domestic)
```
